```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries Loaded Successfully")
```

```python
# --- Load Dataset ---
df = pd.read_csv('../DATA/retail_store_inventory.csv')
#print(df.head())
print(f"Dataset Shape : {df.shape}")
print(f"Dataset Info : {df.info()}")
```

```python
# --- Missing Values ---
print("Missing Values in Each Column")
print(df.isnull().sum())
```

```python
# --- Duplicated Values ---
print("Duplicate Rows")
print(df.duplicated().sum())
```

```python
# --- Basic Statistics ---
print("Basic Statistics")
print(df.describe())
```

```python
# Unique Values
print(f"Unique Categories : {df['Category'].unique()}")
print(f"Unique Regions : {df['Region'].unique()}")
print(f"Unique Stores : {df['Store ID'].unique()}")
print(f"Unique Products : {df['Product ID'].unique()}")
print(f"Weather Conditions : {df['Weather Condition'].unique()}")
print(f"Seasonality : {df['Seasonality'].unique()}")
print(f"Date Range : {df['Date'].min() , df['Date'].max()}")
```

```python
# Convert Date from object to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.strftime('%B')
df['Quarter'] = df['Date'].dt.quarter

print("Date column fixed!")
print(df[['Date', 'Year', 'Month', 'Month_Name', 'Quarter']].head())
```

```python
# --- Revenue ---
df['Revenue'] = df['Units Sold'] * df['Price']
df['Discount_Amount'] = df['Revenue'] * (df['Discount'] / 100)
df['Total_Revenue'] = df['Revenue'] - df['Discount_Amount']

print(df[['Category' , 'Units Sold' , 'Price' , 'Discount' , 'Revenue' , 'Total_Revenue']].head())
```

```python
# --- Discount ---
print(f"Discount Unique Values : {df['Discount'].unique()}")
```

```python
print(f"Dataset Shape After Cleaning : {df.shape}")
print(df.info())
```

```python
# --- Saved Data into new csv --
df.to_csv('../DATA/cleaned_sales_data.csv' , index=False)
print("CSV File Created Succesfully")
```
